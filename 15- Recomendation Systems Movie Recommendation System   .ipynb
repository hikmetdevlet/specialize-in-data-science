### Movie Recommendation System   

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.rename(columns={'movie_id':'id'},inplace=True)
```

```python
movies.head()
```

```python
movies.drop(columns=['homepage','status','production_countries'])
```

```python
#Content Based Recommendation
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
movies['overview'] = movies['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(movies['overview'])
```

```python
print(tfv_matrix)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
indices = pd.Series(movies.index, index=movies['original_title']).drop_duplicates()
```

```python
def give_recomendations(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies['original_title'].iloc[movie_indices]
```

```python
print(give_recomendations('The Dark Knight Rises'))
```

```python
print(give_recomendations('The Godfather'))
```

```python
#save model h5
import pickle
pickle.dump(give_recomendations, open('movie_recomendation.h5', 'wb'))
```
