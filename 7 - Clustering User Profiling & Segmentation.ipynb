```python
#gender mapping
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df.head()
```

```python
df['Income Level'] = df['Income Level'].map({'0-20k': 0, '20k-40k': 1, '40k-60k': 2, '60k-80k': 3, '80k-100k': 4, '100k+': 5})
```

```python
df['Location'].value_counts()
```

```python
df['Location'] = df['Location'].map({'Urban': 0, 'Suburban': 1, 'Rural': 2})
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x=df[['Gender','Income Level','Time Spent Online (hrs/weekend)','Location','Ad Interaction Time (sec)']]
```

```python
model=KMeans(n_clusters=4,random_state=42)
```

```python
model=model.fit(x)
```

```python
tahmin=model.predict(x)
```

```python
tahmin
```

```python
x['cluster']=tahmin
```

```python
silhouette_score(x,tahmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x['Time Spent Online (hrs/weekend)'],x['Ad Interaction Time (sec)'], c=tahmin)
plt.colorbar()
plt.xlabel('Gender')
plt.ylabel('Income_Level')
```

```python
#wcss= within cluster sum of squares
wcss=[]
ss=[]
for i in range(2,10):
    model=KMeans(i)
    model=model.fit(x)
    tahmin=model.predict(x)
    ss1=silhouette_score(x,tahmin)
    ss.append(ss1)
    print(ss1)
    wcss.append(model.inertia_)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=range(2,10),y=wcss,marker='*');
```
